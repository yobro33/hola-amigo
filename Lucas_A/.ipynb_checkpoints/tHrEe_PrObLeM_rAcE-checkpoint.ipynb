```python
# Imports
import numpy
import matplotlib.pyplot as plt

# Initializes
x = []
y = []

# Iterates
for a in range(-500, 500):
    x.append(a)
    y.append(a*a)
    
# Plots
plt.plot(x, y)
```

```python
# Imports
import random

def numberGuess():
    running = True
    lowerBound = input("Lower bound?\n")
    upperBound = input("Upper bound?\n")
    while(running):
        guess = input("Guess?")
        check = random.randint(int(lowerBound), int(upperBound))
        if(guess == check):
            print("good")
        else:
            print("bad")
            
numberGuess()
```
